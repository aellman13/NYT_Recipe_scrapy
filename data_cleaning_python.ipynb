```python
recipes.ingredients = recipes.ingredients.str.lower()
```

```python
recipes.ingredients = recipes.ingredients.apply(lambda x: re.sub(r'[^\w\s]', '', x))
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
print(stop)
```

```python
# Take out all the stop words in each recipe ingredient list
recipes.ingredients = recipes.ingredients.apply(lambda x: ' '.join(x for x in x.split() if x not in stop))
```

```python
# Tokenization 

from textblob import TextBlob
```

```python
#splitting all sentences in to a list of individual words
# TextBlob takes out the periods

recipes.ingredients = recipes.ingredients.apply(lambda x: TextBlob(x).words)
```

```python
# Convert all the individual ingredients back to lists from the textblob object

recipes.ingredients = recipes.ingredients.apply(lambda x: list(x))
```

```python
#Stemming and Lemitization

from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
# converts all the words to the same root
# I didn't apply to this data because it seemed to mess up a lot of words

recipes.ingredients.apply(lambda x: ' '.join([stemmer.stem(word) for word in x]))
```

```python
from nltk import WordNetLemmatizer

lemztr = WordNetLemmatizer()
```

```python
recipes.ingredients = recipes.ingredients.apply(lambda x: ' '.join([lemztr.lemmatize(word) for word in x]))
```
